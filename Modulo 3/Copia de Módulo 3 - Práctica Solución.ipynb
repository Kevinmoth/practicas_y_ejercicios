# Cadena de Caracteres

* Dada una cadena, escriba un programa que cuente cuántas veces aparece una letra específica en la cadena.




```python
cadena = input("Ingrese una cadena: ")
letra = input("Ingrese la letra que desea contar: ")
```

```python
contador = 0
for c in cadena:
    if c == letra:
        contador += 1

print("La letra", letra, "aparece", contador, "veces en la cadena.")
```


* Dada una cadena, escriba un programa que elimine todos los caracteres duplicados y muestre la cadena resultante.

```python
cadena = "aabbcdeeffgh"
```

```python
sin_duplicados = ""
for caracter in cadena:
    if caracter not in sin_duplicados:
        sin_duplicados += caracter

print("La cadena sin duplicados es:", sin_duplicados)
```

```python
numeros = [2, 5, 3, 8, 1, 10, 6]
```

* Dado un diccionario que asocia nombres de frutas con sus calorías, escriba un programa que encuentre la fruta con el menor número de calorías.

```python
frutas_calorias = {"Manzana": 52, "Banana": 89, "Naranja": 47, "Pera": 57}
```

```python
fruta_menor_calorias = None
calorias_menor = float('inf')

for fruta, calorias in frutas_calorias.items():
    if calorias < calorias_menor:
        fruta_menor_calorias = fruta
        calorias_menor = calorias

print("La fruta con el menor número de calorías es:", fruta_menor_calorias)
```

# Problema

Se tiene una lista de nombres de personas con sus respectivas edades, pero algunos de los nombres están mal escritos o en mayúsculas y otros en minúsculas. Además, se tiene un diccionario que asocia nombres y apellidos con sus apodos. Escriba un programa que corrija los nombres mal escritos y agregue los apodos a cada nombre, y luego cree una lista de todas las personas mayores de 18 años y las ordene alfabéticamente por sus nombres completos.

```python
# Lista de personas con sus edades
personas = [
    "Juan Perez, 24",
    "Maria Gonzalez, 18",
    "PEDRO RODRIGUEZ, 32",
    "ana Maria Garcia, 21",
    "Luis ramirez, 16",
    "josefa hernandez, 28"
]

# Diccionario de apodos
apodos = {
    "Juan Perez": "Juancho",
    "Maria Gonzalez": "Mari",
    "Pedro Rodriguez": "Peter",
    "Ana Maria Garcia": "Anita",
    "Luis Ramirez": "Lucho",
    "Josefa Hernandez": "Chefa"
}
```

## Opción 1

```python
def extraer_datos(texto):
  '''
  Esto es un docstring
  texto, es una cadena de caracteres.
  La función devuelve una cadena, con las palabras cada una en mayúscula, usando el método title(), además devuelve la edad, como un valor entero
  Ejemplo:
    extraer_datos("Hola pepito, 25")

  Esto devuelve:
    "Hola Pepito", 25
  '''
  # comentario 2
  datos, edad = texto.split(", ")
  edad = int(edad)
  return datos.title(), edad
```

```python
lista_de_mayores_a_18 = []
lista_de_menores_a_18 = []
for persona in personas:
  datos, edad = extraer_datos(persona)
  # print(datos.capitalize())
  # print(datos.lower())
  # print(datos.upper())
  # print(datos.title())
  if edad > 18:
    print(datos, "-->",edad, "-->", apodos[datos])
    # lista_de_mayores_a_18.append(datos+" ("+apodos[datos]+")")
    lista_de_mayores_a_18.append(f"{datos} ({apodos[datos]})")
  else:
    lista_de_menores_a_18.append(f"{datos} ({apodos[datos]})")
  # print(edad)
lista_de_mayores_a_18.sort()

```

```python
print("Las personas mayores a 18 años son:")
for persona in lista_de_mayores_a_18:
  print("\t",persona)
```

## Opción 2

```python
# Convertir cada elemento de la lista a un diccionario con nombre y edad
personas_dict = {}
for persona in personas:
    nombre, edad = persona.split(", ")
    nombre = nombre.lower().title()  # Convertir el nombre a minúsculas y luego a título (primera letra en mayúsculas)
    edad = int(edad)
    if nombre in apodos:
        nombre_completo = f"{nombre} ({apodos[nombre]})"
    else:
        nombre_completo = nombre
    personas_dict[nombre_completo] = edad

# Crear una lista de personas mayores de 18 años
mayores_de_edad = []
for nombre, edad in personas_dict.items():
    if edad >= 18:
        mayores_de_edad.append(nombre)

# Ordenar la lista alfabéticamente por nombre completo
mayores_de_edad.sort()

# Imprimir la lista de personas mayores de 18 años
print("Personas mayores de 18 años:")
for persona in mayores_de_edad:
    print(persona)
```
